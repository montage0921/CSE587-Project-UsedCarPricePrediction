```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

df=pd.read_csv(r"C:\Users\19692\Downloads\UB CS\2024 Fall\Homework\CES 587\CSE587-Project-UsedCarPricePrediction\carinfo_after_pre_clean.csv")
df.head()
```

# Data Cleaning for feature mileage,brand and price

```python
df=df.dropna(subset=['mileage','price','make','class']) # drop rows with NaN values in mileage, price or make
df = df[df['price']!=0] # drop rows with price equals to 0
```

# Question 1:  How does mileage of a car impact its resale value across different brands? (Te Shi)  
**Hypothesis 1: Generally higher mileage will correlate with a lower resale car value across all car brands**  

```python
# plot relationship between mileage and price
plt.figure(figsize=(10,6))
plt.scatter(df['mileage'],df['price'],alpha=0.5)
plt.title('Mileage vs Price of Used Cars')
plt.xlabel('Mileage (miles)')
plt.ylabel('Price (USD)')
plt.show()
```

```python
# calcualte general correlation value between resale price and mileage
general_corr=df['mileage'].corr(df['price'])
print(general_corr)
```

The General Pearson correlation -0.49 indicates that the mileage has a moderate to fairly strong impacts to a car's final resale.

**Identify Outliers**  
From the graph, although as expect, there is a general trending that with the increase of mileage, the price decreases, some outliers can be identified. Thus linear-regression is used to find cars with abnormal relaitonship between price and mileage.

```python
X = df['mileage'].values.reshape(-1, 1) 


y = df['price']

# create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

df['predicted_price'] = model.predict(X)

df['residuals'] = df['price'] - df['predicted_price']

std_residuals = np.std(df['residuals'])

#outliers are defined as cars that its residual is 3 times or more larger than average residuals
outliers = df[np.abs(df['residuals']) > (3 * std_residuals)]

print(outliers[['mileage', 'price', 'predicted_price', 'residuals']])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(outliers['mileage'], outliers['price'], color='orange', label='Outliers')
plt.scatter(df['mileage'],df['price'],alpha=0.3,label='Normal')
plt.title('Mileage vs Price of Used Cars with Outliers Identified')
plt.xlabel('Mileage (miles)')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
# outliers' correlation value between resale price and mileage
outlier_corr=outliers['mileage'].corr(outliers['price'])
print(outlier_corr)
```

**Hypothesis 2: Although there is a general negative correlation between mileage and resale value, different car brands may have significantly different correlation values**

```python
brand_corr = df.groupby('make')
brand_corr=brand_corr.apply(lambda x: x['mileage'].corr(x['price']))
print(brand_corr)
```

```python
plt.figure(figsize=(10, 6))
brand_corr.plot(kind='bar')

plt.title('Correlation Between Mileage and Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Correlation (Mileage vs Price)')

plt.axhline(general_corr, color='red', linestyle='--', label=f'Average Corr: {general_corr:.2f}')
plt.axhline(outlier_corr, color='green', linestyle='--', label=f'Outlier Corr: {outlier_corr:.2f}')

plt.legend()
plt.grid(True)
plt.show()
```

**Also price drop/mile for each brand is calculated**

```python
def brand_slope(brand_data):
    X = X = brand_data['mileage'].values.reshape(-1, 1) 
    y = brand_data['price']
    model = LinearRegression()
    model.fit(X, y)
    return model.coef_[0]
brand_slopes = df.groupby('make').apply(brand_slope)
print(brand_slopes)
```

```python
plt.figure(figsize=(10, 6))
brand_slopes.plot(kind='bar', color='green')
plt.title('Price Drop per Mile by Brand')
plt.xlabel(
    'Brand')
plt.ylabel('Price Drop per Mile (USD)')
plt.grid(True)
plt.show()
```

# Question 2: Does the popularity of different car classes affect a carâ€™s resale value? (Te Shi)

## More popular car classes will have higher resale values than less popular classes

```python
df = df.groupby('class').filter(lambda x: len(x) > 1)
```

```python
# calculate average and median price for each car class
avg_price_by_class = df.groupby('class')['price'].mean()
avg_price_by_class = avg_price_by_class.reset_index()
avg_price_by_class
```

```python
# for the plot, I would like to show different car_type with different color
# extract car type from class feature
avg_price_by_class['car_type'] = avg_price_by_class['class'].str.split(' - ').str[0]
avg_price_by_class['car_type']

# assign a color for each car type
color_map = {
    'CUV': '#66c2a5',  # Light green
    'Car': '#fc8d62',  # Light orange
    'Pickup': '#8da0cb',  # Light purple
    'SUV': '#e78ac3',  # Light pink
    'Sport Car': '#a6d854',  # Light yellow-green
    'Van': '#ffd92f'  # Light yellow
}

colors=avg_price_by_class['car_type'].map(color_map)
```

```python
# get average price for each type
avg_price_by_type=avg_price_by_class.groupby('car_type')['price'].mean()
print(avg_price_by_type)
```

```python
# Plot the average price by car class
plt.figure(figsize=(10, 6))

plt.bar(avg_price_by_class['class'],avg_price_by_class['price'],color=colors)

for car_type, avg_price in avg_price_by_type.items():
    class_indices = avg_price_by_class[avg_price_by_class['car_type'] == car_type].index

    # crea
    
    plt.hlines(y=avg_price, xmin=class_indices.min(), xmax=class_indices.max(), color='gray', linestyle='--', linewidth=2)

plt.title('Average Resale Price by Car Class')
plt.xlabel('Car Class')
plt.ylabel('Average Resale Price (USD)')
plt.xticks(rotation=90,fontsize=8)
plt.grid(True)
plt.show()
```

### Hypothesis 2: Car class may exhibit differences in the standard deviation of resale prices across different car classes

```python
# calculate the standard deviation for each car class
std_price_by_class = df.groupby('class')['price'].std()
std_price_by_class = std_price_by_class.reset_index(name='std_price')
print(std_price_by_class)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(std_price_by_class['class'], std_price_by_class['std_price'], color='skyblue')
plt.title('Standard Deviation of Resale Prices by Car Class')
plt.xlabel('Car Class')
plt.ylabel('Standard Deviation of Resale Prices (USD)')
plt.xticks(rotation=90, fontsize=8)
plt.show()
```
